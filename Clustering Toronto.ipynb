```python
import pandas as pd
import numpy as np 

import json 

!conda install -c conda-forge geopy --yes 

from geopy.geocoders import Nominatim 

import requests 
from pandas.io.json import json_normalize 

import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans

import folium 
!conda install -c conda-forge folium=0.5.0 --yes
```

```python
# read the tables of giving website to dfs
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)

# we only need the first table, thus, store the first table
table = dfs[0]

# delet the rows that Borough contains 'Not assigned', and store the table into df
df = table[ ~ table['Borough'].str.contains('Not assigned')]
df.head()
```

```python
# check if cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

df.loc[df['Neighborhood'] == 'Not assigned']
```

```python
#  combine rows into one row with the neighborhoods separated with a comma, we define a combine_neighbor function to complete the combination

def combine_neighbor(df):
    return','.join(df.values)

# group the data by Postal Code, and then combine the neighborhoods separated with a comma using defined combine_neighbor function
df_combine = df.groupby(['Postal Code','Borough'])['Neighborhood'].apply(combine_neighbor)

df_combine = pd.DataFrame(df_combine)

# reset the index
df_combine = df_combine.reset_index()

df_combine
```

```python
# check the shape of the new table
df_combine.shape
```
